# ggplot2

### 导入 ggplot2 包

```r
library(ggplot2)
```

### 读取数据

```r
movies <- read.csv("douban_movie_clean.txt", sep="^", header=TRUE)
```

### 用 ggplot2 画出电影时间长度的分布

```r
ggplot(movies) + geom_histogram(aes(x=length))
```
